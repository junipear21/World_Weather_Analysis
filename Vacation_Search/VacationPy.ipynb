```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests

# Import the API key.
from config import g_key

# Configure gmaps API key
gmaps.configure(api_key=g_key)
```

```python
# 1. Import the WeatherPy_database.csv file. 
city_data_df = pd.read_csv("../Weather_Database/WeatherPy_database2.csv")
city_data_df.head()
```

```python
city_data_df.info()
```

```python
# 2. Prompt the user to enter minimum and maximum temperature criteria 
user_min_temp = int(input("What is the minimum temperature you would like?: "))
user_max_temp = int(input("What is the maximum temperature you would like?: "))
```

```python
# 3. Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.
mask = (city_data_df["Max Temp"] <= user_max_temp) & (city_data_df["Max Temp"] >= user_min_temp)
fil_city_data_df = city_data_df.loc[mask]
fil_city_data_df.head(10)
```

```python
# 4a. Determine if there are any empty rows.
fil_city_data_df.isnull().sum()
```

```python
# 4b. Drop any empty rows and create a new DataFrame that doesnâ€™t have empty rows.
df1 = fil_city_data_df.dropna()
df1.head()
```

```python
# 5a. Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = df1[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()

# 5b. Create a new column "Hotel Name"
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# 6a. Set parameters to search for hotels with 5000 meters.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

hotel_names = []

# 6b. Iterate through the hotel DataFrame.
for indx, row in hotel_df.iterrows():
    # 6c. Get latitude and longitude from DataFrame
    target_coordinates = f"{row.Lat}, {row.Lng}"
    params["location"] = target_coordinates
    # 6d. Set up the base URL for the Google Directions API to get JSON data.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # 6e. Make request and retrieve the JSON data from the search. 
    response = requests.get(base_url, params = params)
   
    # 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.
    if (response.status_code == 200):
        places_data = response.json()
        
        if(len(places_data["results"]) > 0): 
            hotel_names.append(places_data["results"][0]["name"])
        else:
            hotel_names.append("")
            
    else:
        hotel_names.append("")
        print("Skipping City")
        
hotel_df["Hotel Name"] = hotel_names
hotel_df.head(10)        
```

```python
# 7. Drop the rows where there is no Hotel Name.
clean_hotel_df = hotel_df.loc[hotel_df["Hotel Name"] != ""]
```

```python
# 8a. Create the output File (CSV)
output_data_file = "hotel_df.csv"
# 8b. Export the City_Data into a csv
clean_hotel_df.to_csv(output_data_file, index_label="City_ID")
```
